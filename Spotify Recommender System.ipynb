```python
listFeat = list(df.columns)
listDataType = []
listNull = []
listNullPct = []
listNunique = []
listUniqSample = []

for item in listFeat:
    listDataType.append(str(df[item].dtype))
    listNull.append(df[item].isna().sum())
    listNullPct.append(round((df[item].isna().sum() / len(df))*100,2))
    listNunique.append(df[item].nunique())
    listUniqSample.append(list(df[item].value_counts().head(2).index))

dictionary = {
    'dataFeature' : listFeat,
    'dataType' : listDataType,
    'null' : listNull,
    'nullPct' : listNullPct,
    'unique' : listNunique,
    'uniqueSample' : listUniqSample
}

pd.DataFrame(dictionary)
```

```python
cb_genres = df[df['popularity'] > 60][['track_name','genre']].reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cb_genres.head()
```

```python
cv = CountVectorizer() 
cv_genres_result = cv.fit_transform(cb_genres['genre'])
```

```python
cv_df = pd.DataFrame(cv_genres_result.todense() ,
                     columns=cv.get_feature_names(),index=cb_genres['track_name'])
cv_df.head()
```

```python
## Measure similarity using cosine similarity
```

```python
cv_df.shape
```

```python
#index_non_duplicates = cb_music_features['track_name'].drop_duplicates().index
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim_genre = cosine_similarity(cv_genres_result)
cos_sim_genre
```

```python
index_to_search = cb_genres[cb_genres['track_name']=='Sanctuary'].index[0]
# index_to_search
index_similar = pd.Series(cos_sim_genre[index_to_search]).sort_values(ascending=False).head(10).index
cb_genres.loc[index_similar]
# cb_genres.iloc[pd.Series(cos_sim_music)[index_to_search].sort_values(ascending=False).head(10).index]
```

```python
def get_recommendation_based_genres(title):
    index_to_search = cb_genres[cb_genres['track_name']==title].index[0]
    index_similar = pd.Series(cos_sim_genre[index_to_search]).sort_values(ascending=False).head(10).index
    return cb_genres.loc[index_similar]
```

```python
get_recommendation_based_genres('Sanctuary')
```

## Artist Recommendation

```python
cb_artist = df[df['popularity'] > 60][['track_name','artist_name']].reset_index(drop=True)
```

```python
cb_artist.head()
```

```python
cb_artist.shape
```

```python
cv = CountVectorizer() 
cv_artist_result = cv.fit_transform(cb_artist['artist_name'])

cos_sim_artist = cosine_similarity(cv_artist_result)
```

```python
def get_recommendation_based_artist(title):
    index_to_search = cb_artist[cb_artist['track_name']==title].index[0]
    index_similar = pd.Series(cos_sim_artist[index_to_search]).sort_values(ascending=False).head(10).index
    return cb_genres.loc[index_similar]
```

```python
get_recommendation_based_artist('Sanctuary')
```

```python
cos_music_features = cosine_similarity(cb_music_features.drop('track_name',axis=1))
```

```python
def get_recommendation_based_features(title):
    index_to_search = cb_music_features[cb_music_features['track_name']==title].index[0]
    index_similar = pd.Series(cos_music_features[index_to_search]).sort_values(ascending=False).head(10).index
    return cb_music_features.loc[index_similar]
```

```python
get_recommendation_based_features('Sanctuary')
```
